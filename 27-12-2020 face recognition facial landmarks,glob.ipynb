```python
import cv2
```

```python
import dlib
```

```python
import imutils
```

```python
from imutils import face_utils
```

#detect the face and then detect the facial landmarks ie 68 points

```python
detector =dlib.get_frontal_face_detector()   # function to detect the face
```

```python
image=cv2.imread('F:/photos/Hardik_Pandya_PTI.jpg')
```

```python
# image,resolution --improve the resolution in the image -- 0 same size
rect=detector(image,1)
```

```python
type(rect)
```

```python
#getting the rect coordinates of 1st face
rect=rect[0]
```

```python
#rect is x1,y1,x2,y2
# rect to x,y,w,h

face_utils.rect_to_bb(rect)
```

```python
x1=rect.bl_corner().x
y1=rect.tl_corner().y
x2=rect.br_corner().x
y2=rect.bl_corner().y
```

```python
image=cv2.rectangle(image,(x1,y1),(x2,y2),(255,0,0),2)
```

```python
cv2.imshow('Image',image)
cv2.waitKey()
```

```python
predictor=dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
```

```python
# image 
# rectangle part

shape=predictor(image,rect)
```

```python
# shape of face to numpy array 
#dlib object is converted to numpy array
facial_landmarks=face_utils.shape_to_np(shape)
```

```python
len(facial_landmarks)
```

```python
for x,y in facial_landmarks:
    image=cv2.circle(image,(x,y),3,(0,255,0),1)
```

```python
cv2.imshow('image',image)
cv2.waitKey()
```

```python
video=cv2.VideoCapture(0)
while True:
    data,frame=video.read()
    rect=detector(frame,1)
    for face in rect:
        shape=predictor(frame,face)
        facial_landmarks=face_utils.shape_to_np(shape)
        for x,y in facial_landmarks:
            cv2.circle(frame,(x,y),2,(0,255,0),2)
    cv2.imshow('Image',frame)
    k=cv2.waitKey(5) & 0xFF
    if k==27:
        break
video.release()
cv2.destroyAllWindows()
```

```python
face_features_model=dlib.face_recognition_model_v1('dlib_face_recognition_resnet_model_v1.dat')
```

```python
face_features_model.compute_face_descriptor(image,shape)
# 128 facial features
```
